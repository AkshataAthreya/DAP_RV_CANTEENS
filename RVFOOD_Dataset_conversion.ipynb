```python
#pandas is used for basic statistical analysis
#numpy used for performing basic operations on array and other mathematical operations. Here we are using it for 
#generating a random value using randint
import pandas as pd
import numpy as np

df = pd.read_csv("DAP food and stuff(responses).csv")
df
```

```python
df.keys()
```

```python
#Here to simplify the dataset for easier analysis we are converting each of the ranges in the coloumns to random values within 
#that range. 

#To perform this we are iterating through each coloumn and checking each element within it and converting it into a whole number
#within the specified range.

#For example here for the coloumn avg spent on breakfast/snacks while iterating through it using a for loop if first element  
#value is lets say 0-50 it will random value from list_0_50 to break_list. simlarly it will keeping appending random values
#in that format to break_list until all elements have been read. 

#We use ranint from random module from numpy which gives a random integer between two values specified

#After this it will reassign the elements of the coloumn to break_list which effectively converts the values in the coloumns to 
#random values without effecting any results.

#Similarly this is done for all the coloumns in the dataset with range values thus cleaning up the data. Once this cleaning 
#up is done the dataframe is then exported back into local directory as a new csv file.

break_list = []
list_0_50 = [20,30,40]
list_50_100 = [50,60,70,80,90]
list_100_150 = [100,110,120,130,140]
list_150_200 = [150,160,170,180,190,200]

for i in df.index:
        if df['How much you spend on average breakfast/snacks daily'][i] == '0-50':
            i = np.random.randint(0,3)
            break_list.append(list_0_50[i])
            
        elif df['How much you spend on average breakfast/snacks daily'][i] == '50-100':
            i = np.random.randint(0,5)
            break_list.append(list_50_100[i])
            
        elif df['How much you spend on average breakfast/snacks daily'][i] == '100-150':
            i = np.random.randint(0,5)
            break_list.append(list_100_150[i])
            
        else:
            i = np.random.randint(0,6)
            break_list.append(list_150_200[i])
            
df['How much you spend on average breakfast/snacks daily'] = break_list
df 
```

```python
break_list = []
list_50_100 = [50,60,70,80,90]
list_100_200 = [100,110,120,130,140,150,160,170,190]
list_200_300 = [200,210,220,230,240,250,260,270,280,290]
list_300_400 = [300,310,320,340,350]

for i in df.index:
        if df['How much  you spend on lunch daily'][i] == '0-100':
            i = np.random.randint(0,5)
            break_list.append(list_50_100[i])
            
        elif df['How much  you spend on lunch daily'][i] == '100-200':
            i = np.random.randint(0,9)
            break_list.append(list_100_200[i])
            
        elif df['How much  you spend on lunch daily'][i] == '200-300':
            i = np.random.randint(0,9)
            break_list.append(list_200_300[i])
            
        else:
            i = np.random.randint(0,5)
            break_list.append(list_300_400[i])
            
df['How much  you spend on lunch daily'] = break_list
df
```

```python
break_list = []

for i in df.index:
        if df['Average waiting time(minutes)'][i] == '0-5':
            break_list.append(np.random.randint(4,6))
            
        elif df['Average waiting time(minutes)'][i] == '5-10':
            break_list.append(np.random.randint(6,11))
            
        elif df['Average waiting time(minutes)'][i] == '10-15':
            break_list.append(np.random.randint(10,16))
            
        else:
            break_list.append(np.random.randint(15,21))
            
df['Average waiting time(minutes)'] = break_list
df
```
